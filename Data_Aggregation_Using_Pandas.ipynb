```python
# Grouped Objects

# When you apply certain  aggregation function to each group seperately then we do grouping

# Number of groups  which is created will be equal to the unique values of the grouping column
```

```python
swiggy_data.head(2)
```

```python
swiggy_data['City'].nunique()
```

```python
# Avg_price - output column name (custom), ('Price'-> Column to agg, np.mean -> func to apply)
# we can inbuilt, third party package function,& userdefined function
# Mulitple  fields can be used to group the dataframe

result_1 = swiggy_data.groupby(['City']).agg(
                                  total_count_order = ('ID', 'count'),
                                  Avg_Price = ('Price', np.mean),
                                  Avg_Delivery_Time = ('Delivery time', np.mean),
                                  Avg_Delivery_Time_hr = ('Delivery time', lambda x:  np.mean(x/60))).reset_index()
```

```python
result_1.loc[result_1['total_count_order'] == result_1['total_count_order'].max()]
```

```python
result_2 = swiggy_data.groupby(['City','Food type']).agg(Count_order = ('ID','count')).reset_index()
```

```python
bangalore_result = result_2.loc[result_2['City'] == 'Bangalore']

bangalore_result.loc[bangalore_result['Count_order'] == bangalore_result['Count_order'].max()]
```

```python
result_2.loc[result_2['City'] == 'Bangalore'].sort_values(by  =['Count_order'], ascending= False)
```
